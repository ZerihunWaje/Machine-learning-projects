```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/ZerihunWaje/Machine-learning-projects/e43c534482e346d335132d58373dfbf998b8d18a/Income%20dataset.csv'
df1 = pd.read_csv(url)
```

```python
df1.head()
```

```python
plt.scatter(df1['Age'],df1['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df1[['Age','Income($)']])
y_predicted
```

```python
df1['cluster']=y_predicted
df1
```

```python
df2=df1[df1.cluster==0]
df3=df1[df1.cluster==1]
df4=df1[df1.cluster==2]
plt.scatter(df2.Age,df2['Income($)'],label='cluster-1',color='green')
plt.scatter(df3.Age,df3['Income($)'],label='cluster-2',color='red')
plt.scatter(df4.Age,df4['Income($)'],label='cluster-3',color='yellow')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
num_scal=df1[['Age','Income($)']]
scaler=MinMaxScaler()
scaler.fit(num_scal)
Scal_val=scaler.transform(num_scal)
scal_df=pd.DataFrame(Scal_val,columns=num_scal.columns)
scal_df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(scal_df)
y_predicted
```

```python
scal_df['cluster']=y_predicted
scal_df
```

```python
km.cluster_centers_
```

```python
df5=scal_df[scal_df.cluster==0]
df6=scal_df[scal_df.cluster==1]
df7=scal_df[scal_df.cluster==2]
plt.scatter(df5.Age,df5['Income($)'],label='cluster-1',color='green')
plt.scatter(df6.Age,df6['Income($)'],label='cluster-2',color='red')
plt.scatter(df7.Age,df7['Income($)'],label='cluster-3',color='yellow')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```
